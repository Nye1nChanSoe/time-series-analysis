#### Lag Features

A lag is just `k steps ago`

- lag_1 = previous observation
- lag_k = value k rows back


#Base on the dataset same lag different meaning-

Hourly data
- lag_1 => 1 hour ago
- lag_12 => 12 hour ago
- lag_24 => same hour yesterday

Daily data
- lag_1 => yesterday
- lag_7 => same day last week

Monthly data
- lag_1 => last month
- lag_12 => same month last year

#### Rolling Window Statistics

Rolling means, Rolling std, min, max over different windows\
They capture recent trends and volatility.

#exponentially weighted moving average - EWMA

- recent points matter more
- smooth
- fast to adapt

`EWMA_t = (alpha * y_t-1) + (1 - alpha) * EWMA_t-1`

EWMA vs Moving Average
- Rapid Regime changes => EWMA
- Stable Seasonality => MA
- Noisy Signals => EWMA
- Long memory => MA
- Short-horizon forecast => EWMA

**Include both** => feature importance => prune
